````python
# Υλοποίηση 1.1

# Αρχικά θέλουμε να εισάγουμε στα δεδομένα μας την πληροφορία της 
# ιθαγένειας για κάθε αντικείμενο της κλάσης Director στη γνώση μας.
# Για να το επιτύχουμε αυτό φορτώνουμε τη γνώση μας στο GraphDB
# και τρέχουμε το παρακάτω SPARQL query:
'''
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ns: <http://example.org/ns/>
PREFIX prop: <http://example.org/prop/>
 
SELECT DISTINCT ?person WHERE { 
    ?person rdf:type ns:Director.
}
'''
# Έχοντας τώρα τη λίστα όλων των ανθρώπων που περιέχονται στη βάση μας μπορούμε
# να εκτελέσουμε queries στο Wikidata ώστε να βρούμε τη χώρα της οποίας την 
# ιθαγένεια έχουν. Η λίστα των ανθρώπων βρίσκεται στο αρχείο 'all_directors.csv'

import pandas as pd
import urllib
from time import sleep

# read list of directors
df = pd.read_csv('all_directors.csv')

directors_countries = []

# for each director find his country of citizenship
for i in range(df.shape[0]):
  # restore name from url
  director_uri = df['person'][i]
  director = director_uri.replace('http://example.org/ns/', '')
  director = urllib.parse.unquote(director)

  # ready wikidata endpoint request
  endpoint = "https://query.wikidata.org/bigdata/namespace/wdq/sparql"
  query = """
  SELECT ?citizenship
  WHERE {
    ?director skos:altLabel|skos:prefLabel|rdfs:label """+'"'+director+'"'+"""@en.
    ?movie wdt:P57 ?director.
    ?director wdt:P27 ?country.
    ?country rdfs:label ?citizenship.
    FILTER(LANG(?citizenship) = "en").
  } LIMIT 1
  """
  r = requests.get(endpoint, params = {'format': 'json', 'query': query})

  # send request and attempt to read response
  try:
    # Parse the answer and return the value
    data = r.json()
    res = []
    if len(data['results']['bindings'])>0:
      for i in range(len(data['results']['bindings'])):
        res.append(data['results']['bindings'][i]['citizenship']['value'])
    else:
      res = None
    directors_countries.append((director_uri, res[0]))
    print(director, ',', res[0])
  except:
    print('Error: ', director, r)

  # wait to avoid Response[429] (too many requests)
  sleep(2)
````

````python
directors_countries = []
with open("directors_countries.txt") as fp:
  line = fp.readline()
  cnt = 1
  while line:
    if cnt%2==1:
      director = line.strip()
    else:
      country = line.strip()
      directors_countries.append((director, country))
    line = fp.readline()
    cnt += 1
````

````python
# Ύστερα εισάγουμε κάθε χώρα στο γράφο γνώσης
# Πρώτα αποθηκεύουμε την πληροφορία σε μορφή n-triples
from rdflib import Graph, Literal, URIRef, Namespace, XSD
import urllib
g = Graph()
ns = Namespace('http://example.org/ns/')
prop = Namespace('http://example.org/prop/')

b = URIRef(prop+'citizen_of')
for (director, country) in directors_countries:
  country_id = urllib.parse.quote(country.encode('utf8'))
  a = URIRef(director)
  c = URIRef(ns+country_id)
  g.add((a,b,c))
g.serialize(destination='directors_countries.nt', format='nt')

# Στη συνέχεια μετατρέπουμε σε owl2
from owlready2 import *
# read n-triples file and convert to owl2
onto = get_ontology('directors_countries.nt').load()
prop = onto.get_namespace('http://example.org/prop/')

with onto:
  with prop:
    class citizen_of(ObjectProperty):
      pass
  #RDF/XML
  onto.save('directors_countries.owl',format='rdfxml')
````

````python
# Προσδιορίζουμε χειροκίνητα τα domain και range της ιδιότητας citizen_of
# με τις παρακάτω γραμμές
"""
<owl:ObjectProperty rdf:about="http://example.org/prop/citizen_of">
    <rdfs:domain rdf:resource="http://example.org/ns/Person"/>
    <rdfs:range rdf:resource="http://example.org/ns/Place"/>
</owl:ObjectProperty>
"""
# Επιπλέον, κάνουμε την ιδιότητα belongs_in transitive ώστε όποια ιδιότητα
# αποδίδουμε σε κάποιο Place να ισχύει και για τα Place στα οποία εντάσσεται.
# Αυτό το επιτυγχάνουμε με τον παρακάτω τρόπο
"""
<owl:TransitiveProperty
    rdf:about="http://example.org/prop/belongs_in">
</owl:TransitiveProperty>
"""
# Τώρα φορτώνουμε την νέα εμπλουτισμένη γνώση στο GraphDB όπου και εκτελούμε
# το παρακάτω SPARQL ερώτημα:
"""
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX ns: <http://example.org/ns/>
PREFIX prop: <http://example.org/prop/>
 
SELECT DISTINCT ?place WHERE { 
    ?place rdf:type ns:Place.
}
"""
# και αποθηκεύουμε τα αποτελέσματα στο αρχείο 'all_places.csv'. Στη συνέχεια 
# για κάθε μέρος που επιστράφηκε εκτελούμε ένα SPARQL query για να βρούμε το 
# Geonames ID του το οποίο και χρησιμοποιούμε για να βρούμε ιεραρχικά τις 
# περιοχές στις οποίες ανήκει το κάθε μέρος που βρήκαμε. Την πληροφορία αυτή
# σώζουμε ξανά στο γράφο γνώσης μας.
import requests
import pandas as pd
import urllib
from time import sleep

# read list of directors
df = pd.read_csv('all_places.csv')

places_ids = []

# for each place find its Geonames ID
for i in range(df.shape[0]):
  # restore name from url
  place_uri = df['place'][i]
  place = place_uri.replace('http://example.org/ns/', '')
  place = urllib.parse.unquote(place)

  # ready wikidata endpoint request
  endpoint = "https://query.wikidata.org/bigdata/namespace/wdq/sparql"
  query = """
  SELECT ?geo_id
  WHERE {
    ?place skos:altLabel|skos:prefLabel|rdfs:label """+'"'+place+'"'+"""@en;
    wdt:P1566 ?geo_id.
  } LIMIT 1
  """
  r = requests.get(endpoint, params = {'format': 'json', 'query': query})

  # send request and attempt to read response
  try:
    # Parse the answer and return the value
    data = r.json()
    res = []
    if len(data['results']['bindings'])>0:
      for i in range(len(data['results']['bindings'])):
        res.append(data['results']['bindings'][i]['geo_id']['value'])
    else:
      res = None
    places_ids.append((place_uri, res[0]))
    print(place, ',', res[0])
  except:
    print('Error: ', place, r)

  # wait to avoid Response[429] (too many requests)
  sleep(2)
````

Απάντηση 1.1.1<br>
Η πληροφορία για την ιεραρχία των αντικειμένων της κλάσης Place μοντελοποιείται με χρήση της ιδιότητας 'belongs_in'. Η ιδιότητα αυτή έχει Domain και Range την κλάση Place, όπου για κάθε τριάδα της μορφής place1 prop:belongs\_in place2 το place2 βρίσκεται σε υψηλότερη ιεραρχικά θέση από το place1 (π.χ. Greece prop:belongs\_in Europe).<br>
Η ιδιότητα belongs\_in είναι μεταβατική (transitive), καθώς αν για παράδειγμα το Los Angeles ανήκει στην California και η California ανήκει στις ΗΠΑ, τότε και το Los Angeles ανήκει στις ΗΠΑ.<br>
Επιλέξαμε συνειδητά να μην εισάγουμε την ιεραρχία αυτή και στις άλλες ιδιότητες της κλάσης Place όπως για παράδειγμα τη περιοχή που γυρίστηκε μια ταινία, γιατί πρόκειται για μια πιο ειδική πληροφορία. Αν θέλουμε να βρούμε γενικότερες πληροφορίες μπορούμε να χρησιμοποιήσουμε την ιδιότητα belongs\_in στο SPARQL query μας.

Τα κείμενα που θέλουμε να αναλύσουμε (είτε οι περιγραφές είτε τα σχόλια των χρηστών) μπορεί να περιέχουν και πληροφορία την όποια δεν χρειαζόμαστε, όπως html tags ή χαρακτήρες όπως παρενθέσεις, brackets κ.ά., οι οποίοι δεν προσδίδουν κάποια πληροφορία. Για αυτό πριν την ταξινόμηση των κειμένων οι χαρακτήρες αυτοί πρέπει να αφαιρεθούν. 

Για την αφαίρεση των html tag χρησιμοποιούμε την βιβλιοθήκη **bs4** που περιέχει την συνάρτηση **BeautifulSoup**, ενώ για την αφαίρεση των υπόλοιπων χαρακτήρων έχουμε κατασκευάσει εμείς αντίστοιχες συναρτήσεις. 

Για παράδειγμα στο δικό μας Task δεν μας ενδιαφέρει αν τα γράμματα είναι πεζά ή κεφαλαία όποτε επιλέγουμε εμείς να γίνουν όλα πεζά. Επίσης μπορεί να μην μας ενδιαφέρει το περιεχόμενο μέσα σε παρενθέσεις ή αγκύλες όποτε και αυτό μπορούμε αν θέλουμε να το αφαιρέσουμε με την συνάρτηση `remove_between_square_brackets()`. Το ίδιο και με τους ειδικούς χαρακτήρες ή αριθμούς μέσω της συνάρτησης `remove_special_characters()`.



````python
#Removing the html strips
def strip_html(text):
    '''
    Συνάρτηση αλλάξει μετατρέπει αφαιρεί το html περιεχόμενο μέσα από το κείμενο
    '''
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()


def lower_text(text):
    '''
    Συνάρτηση αλλάξει μετατρέπει όλους τους χαρακτηρες σε πεζα
    '''
    return text.lower()

def remove_between_square_brackets(text):
    '''
    Συνάρτηση που αφαιρεί το περιεχόμενο που βρίσκεται εντός bracket
    '''
    return re.sub('\[[^]]*\]', '', text)

def remove_special_characters(text, remove_digits=True):
    '''
    Συνάρτηση που αφαιρεί ειδικούς χαρακτήρες ή και τα γραμματα αν θέλουμε
    '''
    if remove_digits:
      pattern=r'[^a-zA-z\s]'
    else:
      pattern=r'[^a-zA-z0-9\s]'
    text=re.sub(pattern,'',text)
    return text
````

Αυτά αφορούν την αφαίρεση των λέξεων που δεν περιέχουν ιδιαίτερη πληροφορία όπως οι συνδετικές λέξεις και την κανονικοποιήση των λέξεων σε μια βασική μορφή με την αφαίρεση των καταλήξεων τους.

Για την αφαίρεση των stopwords θα χρησιμοποιήσουμε την βιβλιοθήκη nltk η οποία μας παρέχει μια πλήρη λίστα με όλες αυτές τις λέξεις για διάφορες γλώσσες. Τέλος για την κανονικοποιήση των λέξεων παρέχουμε 2 συναρτήσεις που χρησιμοποιούν την PorterStemmer και WordNetLemmatizer της ίδια βιβλιοθήκης. 

Περισσότερες πληροφορίες για την λειτουργία των μεθόδων αυτών μπορείτε να βρείτε [εδώ](https://www.tutorialspoint.com/natural_language_toolkit/natural_language_toolkit_stemming_lemmatization.htm).


````python
tokenizer=ToktokTokenizer() #για σπάσουμε το κείμενο σε tokens

stopword_list=nltk.corpus.stopwords.words('english') # η λίστα των stopwords

#removing the stopwords
def remove_stopwords(text, is_lower_case=False):
    tokens = tokenizer.tokenize(text)
    tokens = [token.strip() for token in tokens]
    if is_lower_case:
        filtered_tokens = [token for token in tokens if token not in stopword_list]
    else:
        filtered_tokens = [token for token in tokens if token.lower() not in stopword_list]
    filtered_text = ' '.join(filtered_tokens)    
    return filtered_text


def stemmer(text):
    words = nltk.word_tokenize(text)
    ps=nltk.porter.PorterStemmer()
    text= ' '.join([ps.stem(word) for word in words])
    return text


lm = WordNetLemmatizer()
def lemmatizer(text):
    words = nltk.word_tokenize(text)
    text= ' '.join([lm.lemmatize(word) for word in words])
    return text

# remove single letters as no valuable info is conveyed
def letter_remover(text):
  words = nltk.word_tokenize(text)
  text = ' '.join([word for word in words if len(word)>1])
  return text
````

Όλες οι παραπάνω συναρτήσεις δέχονται σαν είσοδο ένα text και επιστρέφουν ένα text. Έτσι με τον τρόπο αυτό μπορούμε να κατασκευάσουμε μια αλυσίδα ενεργειών που θέλουμε να εφαρμόσουμε σε ένα κείμενο. H παρακάτω συνάρτηση δέχεται σαν είσοδο ένα κείμενο και μια λίστα με συναρτήσεις και τις καλεί σειριακά δημιουργώντας ένα pipeline προ-επεξεργασίας του κειμένου.

````python
def preprocess_text(text, pipeline):
    '''
    Η συνάρτηση η οποία καλεί όλες τις μεθόδους του pipeline ώστε να γίνει
    η σωστή προεπεξεργασία του κειμένου εισόδου
    '''
    for method in pipeline:
        text = method(text)
    return text
````

Όπως είναι λογικό, η καλύτερη μελέτη των περιγραφών θα οδηγήσει σε καλύτερες ομαδοποιήσεις.  

Παρακάτω ορίζουμε ένα τέτοιο βασικό σύνολο για την θεματική: Πολεμικές ταινίες.

Σημαντικό σε αυτό το σημείο είναι να περάσουμε και τα keywords από το pipeline μας ώστε και αυτές να είναι στην ίδια βασική δομή.

````python
# match 1 keyword
def find_group_1(keywords, pipeline = pipeline):
    group_id = []
    for imdb_id, description in descriptions.items():
        description = preprocess_text(description, pipeline)
        for keyword in keywords:
            if keyword in description:
                group_id.append(imdb_id)
    return list(set(group_id))

# match 2 keywords
def find_group_2(keywords, pipeline = pipeline):
    group_id = []
    for imdb_id, description in descriptions.items():
        description = preprocess_text(description, pipeline)
        for keyword in keywords:
            if keyword in description:
                for keyword2 in keywords:
                    if keyword2 != keyword and keyword2 in description:
                        group_id.append(imdb_id)
    return list(set(group_id))
````

````python
reviews = {}
pipeline = [strip_html,
            remove_between_square_brackets,
            lower_text,
            remove_special_characters,
            remove_stopwords,
            lemmatizer,
            letter_remover,
]

for user_id, imdb_id, comment in zip(movies_comments["user_id"],
                                     movies_comments["imdb_title_id"],
                                     movies_comments["comment"]):
  reviews[str(user_id)+"_"+imdb_id] = comment

print(reviews)
````

````python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
positive = []
negative = []
pos_comments = []
neg_comments = []

for user_review in reviews:
  user, imdb_id = user_review.split('_')

  comment = reviews[user_review]
  comment = preprocess_text(comment, pipeline)
  score = sia.polarity_scores(comment)

  if score["compound"] > 0:
    positive.append((user, imdb_id))
    pos_comments.append(comment)
  else:
    negative.append((user, imdb_id))
    neg_comments.append(comment)
````
